```python
import pandas as pd 
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("Per Capita Income (US$)")
plt.scatter(df[["year"]], df[["per capita income (US$)"]], color = "green")
```

```python
new_df = df.drop(df[["per capita income (US$)"]], axis = "columns")
new_df
```

```python
income = df[["per capita income (US$)"]]
income
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df, income)
```

```python
reg.predict([[2020]])
```

```python
predictions_df = pd.read_csv("predictions.csv")
predictions_df
```

```python
i = reg.predict(predictions_df)
i
```
